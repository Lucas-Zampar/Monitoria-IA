```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 2, 100)
noise = np.random.normal(0, .1, x.shape)
y = -x**3 + 2*x**2 + noise

y0 = x
y1 = -3*x + 6
y3 = -x**3 + 2*x**2

plt.xlim(0,2.5)
plt.ylim(-0.5, 2)

plt.scatter(x,y)
plt.plot(x,y0, linewidth=3, color='red')
plt.plot(x,y1, linewidth=3, color='orange')
plt.plot(x,y3, linewidth=6, color='green')
```

```python
def softmax(vetor):
    d = np.exp(vetor).sum()
    return np.array([np.exp(_)/d for _ in vetor])

x = 3*np.random.rand(100)
x0 = softmax(x)


plt.scatter(x, x0, color='g')
```
